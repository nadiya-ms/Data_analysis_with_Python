# Python Dictionary Tips and Tricks 

### Using .get() Instead of [] for Lookups
If you are working with a dictionary and you need to access a value but the key is not there use the **.get()** method. It's safer and you can set a default value if the key is missing. However, if a missing key is a bug, use square brackets. You want the error to show up right away in that case.

```python
config = {"debug": True, "verbose": False}
print(config.get("timeout", 30))
```

### Using defaultdict for Grouping Data
If you're working with a list of words and you want to count how many times each word appears, you might code it like this:

```python
words = ["apple", "banana", "apple", "cherry", "banana", "banana"]

counts = {}

for word in words:
    if word not in counts:
        counts[word] = 0
    counts[word] += 1

print(counts)
```

This works, but it's a bit verbose. Python's **defaultdict** makes it cleaner. With **defaultdict(int)**, Python automatically creates a default value of 0 whenever a missing key is accessed.

```python
from collections import defaultdict
words = ["apple", "banana", "apple", "cherry", "banana", "banana"]

counts = defaultdict(int)
for word in words:
    counts[word] += 1
print(counts)
```

More Pythonic alternatives - Using dict.get()

```python
words = ["apple", "banana", "apple", "cherry", "banana", "banana"]
counts = {}
for word in words:
    counts[word] = counts.get(word, 0) + 1
print(counts)
```

### Merging Dictionaries With the | Operator
In modern Python, the cleanest way to merge dictionaries is with the | operator.

```python
defaults = {"color": "blue", "size": "medium"}
overrides = {"size": "large", "weight": "heavy"}

merged = defaults | overrides
print(merged)
```

When keys overlap, the dictionary on the right side wins. If you want to do in-place merging, you can use the **|=** operator:

```python
defaults |= overrides
print(defaults)
```

### Unpacking Dictionaries into Function Arguments
If you have a function and a dictionary, and their fields or keys match. Instead of passing the keys one by one, like `name=data["name"]`, `age=data["age"]`, you can pass everything using the `**` double-asterisk operator. Let's create a user function and some dummy user data to understand it:

```python
def create_user(name, age, role="viewer"):
    return {"name": name, "age": age, "role": role}

user_data = {
    "name": "David",
    "age": 33,
    "role": "editor"
}
```

```python
# Normal Way
user = create_user(
    name=user_data["name"],
    age=user_data["age"],
    role=user_data["role"]
)

print(user)
```

```python
# Using **
print(create_user(**user_data))
```

### Using the Walrus Operator With Dicts
Python 3.8 introduced the walrus operator **(:=)**, which lets you assign a value as part of an expression. This is really useful with dictionaries.

Let's say you have a dictionary and you want to get the user data and their name if they exist. This is typically how you would normally code it:

```python
data = {
    "user": {
        "name": "Bryan",
        "email": "bryan@gmail.com"
    }
}

if data.get("user") is not None:
    user = data.get("user")
    name = user.get("name")

    print(name)
```

This works, but it repeats the same dictionary lookup multiple times. You can replace it with the walrus operator (:=), which looks up and assigns the value in a single step. This is especially helpful when working with nested dictionary structures.

```python
if (user := data.get("user")) is not None:
    name = user.get("name")

    print(name)
```

### Using **TypedDict** for Structured Data

```python
def greet(user):
    return f"Hello, {user['name']}!"

user = {
    "name": "Clair",
    "age": "thirty"
}

print(greet(user))
```

This works at runtime, but there is a hidden problem: "age" is supposed to be a number, not a string. Python itself will not complain, which can lead to bugs later in larger projects. TypedDict makes the expected dictionary structure explicit:

```python
from typing import TypedDict

class UserProfile(TypedDict):
    name: str
    age: int

def greet(user: UserProfile) -> str:
    return f"Hello, {user['name']}!"
```

Now tools like mypy can catch mistakes before the code runs:

```python
user: UserProfile = {
    "name": "Clair",
    "age": "thirty",
}

print(greet(user))
```

### Iterating Easily: **.items(), .keys(), .values()**
Python dictionaries have many built-in methods for iteration: `.items(), .keys(), and .values()`. Most developers know about them, but don't use them as often as they should. They might loop over a dictionary like this:

```python
scores = {
    "David": 92,
    "Bryan": 87,
    "Clair": 95
}

for name in scores:
    print(name, scores[name])
```

That works. But it's not the best way — it does an extra dictionary lookup every time through the loop. Python's `.items()` method is cleaner:

```python
for name, score in scores.items():
    print(name, score)
```

```python
for name in scores.keys():
    print(name, score)
```

```python
for score in scores.values():
    print(name, score)
```

### {} dictionary comprehension

```python
{k: v for k, v in [(1, 2), (3, 4)]}
```

```python
{n: n for n in range(5)}
```

```python
{chr(n): n for n in (65, 66, 66)}
```

```python
{k: v for k, v in (('I', 1), ('II', 2))}
```

```python
{k: v for k, v in (('a', 0), ('b', 1)) if v == 1}
```
