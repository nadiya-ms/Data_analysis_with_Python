## Cycle **while**

```python
x = 0
while x != 5:
    x += 1
    print(x)
    break
    
print('I stopped working')
```

```python
x = 5
while x != 0:
    if x % 2 == 0:
        print(x, '- even number')
    else:
        print(x, '- odd number')
    x = x - 1
```

## Cycle **for**

```python
# iterating through a string
company_name = 'Orange'
for letter in company_name: 
    letter = letter.capitalize() 
    print(letter)
```

## break, pass, continue

```python
phrase = 'We all studied a little—something or other, and somehow or other. — A.S. Pushkin'
```

```python
for letter in phrase:
    if letter == ' ':
        break
    print(letter, end='')
```

```python
for letter in phrase:
    if letter == ' ':
        continue
    print(letter)
print('finish loop')
```

```python
for letter in phrase:
    if letter == ' ':
        pass
    print(letter)
print('finish loop')
```

## List comprehension

```python
# Given a sequence of numbers, we want to keep only those that are divisible by 5.
sequence = range(0, 40, 3)
list(sequence)
```

```python
# simple way to do it
for num in sequence:
    if num % 5 == 0:
        print(num)
```

```python
# if we want to save the result in a list, we can do it like this
filtered_sequence = []

for num in sequence:
    if num % 5 == 0:
        filtered_sequence.append(num)

print(filtered_sequence)
```

```python
[num for num in sequence if num % 5 == 0]
```

```python
[w for w in sequence if w % 5 == 0]
```

```python
# with list comprehension it's shorter

[x**2 for x in sequence if x % 5 == 0]
```
