## Text Formatting

To format text in *italics*, enclose it within asterisks (*) or underscores—_like this_ or __like this__ or *like this*.

To format text in **bold**, enclose it within double asterisks (**) — **like this**.

Alternative methods allow for simultaneous formatting within lists:
_text can be both **bold** and italic_

## Lists

To create an unordered list, prefix each item with an asterisk (*), like this:
* Item 1
* Item 2
* Item 3

To create an ordered list, simply enter the numbering:
1. Item 1
2. Item 2
3. Item 3
