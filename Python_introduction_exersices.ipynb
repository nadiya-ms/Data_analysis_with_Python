3. Display a list of the unique values ​​from the input list. Input data: 4 8 6 2 1 2 3 4 4 0 2 4 4 5 8 2 9 6 0 0

```python
unique_numbers = []
numbers = map(int, input().split())
for number in numbers:
    if number not in unique_numbers:
        unique_numbers.append(number)
print(unique_numbers)

for number in unique_numbers:
    print(number, end=' ')
```

5. Write a function that accepts a number `n` as a parameter and displays the first `n` Fibonacci numbers, excluding 0. Input: 3.

```python
def fibonacci(n):
    fib_sequence = []
    a, b = 0, 1
    while len(fib_sequence) < n + 1:
        fib_sequence.append(a)
        a, b = b, a + b
    return fib_sequence

n = int(input())
fib_list = fibonacci(n)
fib_string = ' '.join(map(str, fib_list ))

print(fib_string)
```

С помощью анонимной lambda-функции создайте столбец "EurGrad", который переводит значения оценок столбца 'grade'  в европейскую систему. 
Входные данные:
stud_id;Subject;grade
1;ТВиМС;60
2;МОР;70
3;ТВиМС;80
4;МОР;85
5;ТВиМС;100
Выходные данные:
E D C B A

```python
import pandas as pd
import io
import sys

header = sys.stdin.readline()

grade_to_eur = lambda x: (
    'A' if x > 90 else
    'B' if x > 80 else
    'C' if x > 70 else
    'D' if x > 60 else
    'E'
)

eur_grades = []

for line in sys.stdin:
    parts = line.strip().split(';')
    if len(parts) > 2:
        grade = int(parts[2].strip())
        eur_grades.append(grade_to_eur(grade))

print(" ".join(eur_grades))
```
